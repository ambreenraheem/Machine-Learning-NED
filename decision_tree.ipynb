````python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import datasets
import graphviz

# Dataset load karo (example: Iris)
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Model banao
clf = DecisionTreeClassifier(
    max_depth=3,          # tree kitna deep ho
    criterion='gini',     # split ka metric (gini ya entropy)
    min_samples_split=2   # split ke liye min samples
)

# Train karo
clf.fit(X, y)

# ── Graphviz se visualize karo ──
dot_data = export_graphviz(
    clf,
    out_file=None,              # file nahi, string return karo
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,                # nodes color se fill honge
    rounded=True,               # rounded boxes
    special_characters=True
)

# Graph render karo
graph = graphviz.Source(dot_data)
graph.render("decision_tree", format="png", cleanup=True)  # PNG save hoga
graph.view()                                               # directly open bhi ho ga



# **Graphviz alag se kya dikhata hai? Har node mein yeh info hoti hai:**
# ```
# sepal length <= 2.45      ← split condition
# gini = 0.667              ← impurity
# samples = 150             ← kitne data points aye
# value = [50, 50, 50]      ← har class ke samples
# class = setosa            ← majority class
````

what is gini
In the context of the decision tree, 'gini' refers to the Gini impurity. It's a metric used to measure the impurity or disorder of a node in a decision tree. A lower Gini impurity indicates a more homogeneous node, meaning most samples in that node belong to the same class. When building the tree, the algorithm tries to find splits that minimize the Gini impurity in the resulting child nodes.


Explain the nodes in the decision tree visualization
Each node in the decision tree visualization provides the following information:

split condition (e.g., sepal length <= 2.45): This is the condition used to split the data at that node. If the condition is true, the data goes to the left child node; otherwise, it goes to the right child node.
gini (e.g., gini = 0.667): This is the Gini impurity of the node. It measures the randomness or impurity of the samples at that node. A value of 0 means the node is perfectly pure (all samples belong to the same class).
samples (e.g., samples = 150): This indicates the number of data points (samples) that reached this node.
value (e.g., value = [50, 50, 50]): This array shows the count of samples for each class within that node. For the Iris dataset, [50, 50, 50] means there are 50 samples of class 0, 50 of class 1, and 50 of class 2.
class (e.g., class = setosa): This is the majority class of the samples in that node. If the node were a leaf node, this would be its predicted class.

so gini is the model
No, gini is not the model itself. It is a criterion or a metric that the DecisionTreeClassifier (which is the model) uses to decide how to split the data at each node. Specifically, it measures the 'impurity' of a node, and the algorithm tries to find splits that minimize this impurity.
